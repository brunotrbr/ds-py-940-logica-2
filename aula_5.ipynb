## Exercício 1 - 940-ds-py-dicionarios
Crie uma agenda telefônica no console para que seja possível adicionar contatos, contendo as seguintes características

Contato

Nome do contato
Telefone
Email
Na agenda deve ser possível:

Adicionar um contato
Remover um contato
Listar todos os contatos (trazendo todos os dados do contato)
Listar todos os contatos de forma resumida (trazendo somente o nome do contato)
Buscar contatos por nome
Obs:

Para armazenar os dados no programa, pode ser utilizada um dicionário.

```python
from typing import Dict

contatos: Dict[str,Dict] = {}

def cadastro():
    nome = input('Digite o nome completo do contato: ')
    telefone = input(f'Digite o telefone do contato {nome}: ')
    email = input('Digite o email do contato: ')
    # Aqui temos que tratar a possibilidade de existir dois contatos com o mesmo nome.
    # Opção 1: Confirma a sobrescrita do contato
    # Opção 2: Alterar nome (copy)
    # Opção 3: Mudar a chave do dicionário

    # Aqui vamos manter sobrescrevendo.
    contatos[nome] = {'nome': nome, 'telefone': telefone, 'email': email}


while True:
    escolha = input('\n-----------------------------MENU----------------------------\n1 para cadastrar contatos,'
                        '\n2 para listar todos os contatos,\n3 para listar os contatos resumidamente,'
                        '\n4 para consultar contatos por nome,\n5 para apagar contato'
                        '\n0 para sair\n-->: ')

    if escolha == '0':
        print('Programa encerrado.')
        break
    elif escolha == '1':
        cadastro()

    elif escolha == '2':
        if not contatos:
            print('\nNenhum contato cadastrado. Retornando ao Menu...')
        else:
            print('\nOs contatos cadastrados são:\n')
            for i in sorted(contatos, key=lambda contato: contato[1]):
                print(contatos[i])
    else:
        print("Opção Inválida.")
```
